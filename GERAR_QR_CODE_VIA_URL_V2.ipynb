```python
# Instalar a biblioteca necessária
!pip install qrcode[pil]

# Importar as bibliotecas
import qrcode
from IPython.display import display

# URL que você quer transformar em QR Code
url = "https://sougov.sigepe.gov.br/sougov/"

# Criar o QR Code
qr = qrcode.make(url)

# Exibir o QR Code
display(qr)

# Salvar a imagem (opcional)
qr.save("qrcode.png")
```
